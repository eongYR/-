```python
import math

#####################(INPUT)#####################

CLO = 0.5  #50  (Clothing)
MET = 1.0  #60  (Metabolic Rate)
WME = 0  #70  (Extercal Work(단, Normally Around "0"))
TA = 26  #80  (Air Temperature)
TR = 26  #90  (Mean Radiant Temperature)
VEL = 0.10  #100 (Relative Air Velocity)
RH = 0.50  #120 (Relative Humidity)
PA = 0  #130 (Water Vapour Pressure)
#####################(INPUT)#####################

FNPS = math.exp(16.6536 - 4030.183 / ( TA + 235))  #140
if PA == 0:
    PA = RH * 10 * FNPS  #150 WVP Method / Water Vapour Pressure

M2 = RH * FNPS * 1000
ICL = 0.155 * CLO  #160  Thermal Insulation of the Clothing
M = MET * 58.15  #170  Metabolic Rate
W = WME * 58.15  #180  External Work
MW = M - W  #190 Internal Heat Production in the Human
FCL = ICL #200
if ICL <= 0.078:
    FCL = 1 + 1.29 * ICL
else:
    FCL = 1.05 + 0.645 * ICL

HCF = 12.1 * math.sqrt(VEL) #210
TAA = TA + 273 #220
TRA = TR + 273 #230
TCLA = TAA + (35.5 - TA) % (3.5 * ICL + 0.1) #250
P1 = ICL * FCL #260
P2 = P1 * 3.96 #270
P3 = P1 * 100  #280
P4 = (P1 * TAA) #290
P5 = 308.7 - 0.028 * MW + P2 * (TRA / 100) ** 4 #300
XN = TCLA / 100 #310
XF = XN #320
HCN = 2.38 * abs(100 * XF - TAA) ** 0.25
if HCF > HCN:
  HC = HCF
else:
  HC = HCN

EPS = 0.00015
for _ in range(149):
    XN = (P5 + P4 * HC - P2 * XF ** 4) / (100 + P3 * HC)
    if abs(XF - XN) > EPS:
        XF = (XF + XN) / 2
    else:
        break

TCL = 100 * XN - 273
HL1 = 3.05 * 0.001 * (5733 - 6.99 * MW - M2)
HL2 = 0.42 * (M - 58.15)


HL3 = 1.7 * 0.00001 * M * (5867 - M2)
HL4 = 0.0014 * M * (34 - TA)
HL5 = 3.96 * FCL * ((XN**4) - (TRA / 100) ** 4)
HL6 = FCL * HC * (TCL - TA)
TS = 0.303 * math.exp(-0.036 * M) + 0.028
PMV = TS * (MW - HL1 - HL2 - HL3 - HL4 - HL5 - HL6)
PPD = 100 - 95 * math.exp(-0.03353 * PMV ** 4 -0.2179 * PMV ** 2)
print("******************************************")

print("INPUT")
print("CLO:", CLO)
print("MET:", MET)
print("WME:", WME)
print("TA:", P4)
print("VEL:", VEL)
print("RH:", RH)
print("PA:", PA)

print("******************************************")

print("******************************************")
print("Calculation Logic")
print("FNPS:", FNPS)
print("M2(Water Vapour Pressure):", M2)
print("ICL:", ICL)
print("M:", M)
print("W:", W)
print("MW:", MW)
print("FCL:", FCL)
print("TAA:", TAA)
print("TRA:", TRA)
print("HCF:", HCF)

print("******************************************")
print("P값")
print("P1:", P1)
print("P2:", P2)
print("P3:", P3)
print("P4:", P4)
print("P5:", P5)
print("******************************************")

print("******************************************")
print("HC:", HC)
print("XN:", XN)
print("TCL:", TCL)
print("******************************************")

print("******************************************")
print("HL1:", HL1)
print("HL2:", HL2)
print("HL3:", HL3)
print("HL4:", HL4)
print("HL5:", HL5)
print("HL6:", HL6)
print("******************************************")

print("******************************************")
print("TS(Thermal Sensation Trans. Coeff.):", TS)
print("predicted Mean Vote (PMV):", PMV)
print("PPD:", PPD)
print("******************************************")


```
